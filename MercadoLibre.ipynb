# WebScrapping Mercado Libre (Precio de pantallas)

### Importaciones necesarias

```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
```

### Consulta de la pagina de busqueda y verificacion de resultados con la busqueda de las meta de las pagina

```python
response = urlopen("https://listado.mercadolibre.com.mx/television-40-pulgadas#D[A:television%2040%20pulgadas]")

soup = BeautifulSoup(response.read(), 'html.parser')

metadata = soup.find_all('meta')
print(metadata)
```

### Variables necesarias

```python
listProd = {
    "Nombre": [],
    "Precio": []
}
```

### Se consultan la lista de contenedores de las pantallas resultantes buscadas

```python
listProduc = soup.select('.ui-search-layout__item')
```

### Metodo para generar diccionarios con la informacion de las pantallas y mostrarlas en Pandas

```python

for i in listProduc:
    listProd["Nombre"].append(i.select('.ui-search-item__title')[0].contents[0].text)
    listProd["Precio"].append(i.select('.price-tag-fraction')[0].contents[0].text)
    
    
dfP = pd.DataFrame(listProd)
dfP
    
```

### Ordenar valores de la mas barata a la mas cara

```python
precioDF = dfP.sort_values('Precio')
precioDF
```
